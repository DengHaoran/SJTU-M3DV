# 数据简介
这是一个关于肺部结节的CT扫描片的数据集

```python
import numpy as np
from matplotlib import pyplot as plt
```

每一个结节是一个数据点，保存在一个*npz*文件中，我们先把它们读进来。

```python
tmp = np.load('./nodule/candidate1.npz')
```

一个npz文件是一个病灶，含有两个3D的**NumPy Array**,分别是‘voxel’和‘seg’。

voxel保存的是CT扫描的数据，seg是一个布尔array，标False的地方不是结节，标True的地方是结节(seg的作用是把CT中的结节给抠出来)。

```python
tmp['voxel']
```

```python
tmp['seg']
```

```python
np.shape(tmp['voxel'])
```

```python
np.shape(tmp['seg'])
```

```python
voxel = tmp['voxel']/255
seg = tmp['seg']
```

为了让大家对这两个数据有更加深刻的认识，下面是对于一个病灶的逐层的可视化，左边是seg，标出结节的位置，右边是CT的样子。

```python
for i in range(37,60,2):
    plt.figure()
    plt.subplot(1,2,1)
    plt.imshow(seg[i])
    plt.subplot(1,2,2)
    plt.imshow(voxel[i])
    plt.show()
```
